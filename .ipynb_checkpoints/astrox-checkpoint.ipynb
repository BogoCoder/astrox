```python
def votable_to_pandas(votable_file):
    votable = parse(votable_file)
    table = votable.get_first_table().to_table(use_names_over_ids=True)
    return table.to_pandas()
```

#### K-Means

```python
features_coord = ['ra', 'dec','hard_hm','hard_hs', 'hard_ms', 'var_prob_b', 'var_sigma_b',
            'var_prob_h', 'var_sigma_h',
        'var_prob_m', 'var_sigma_m', 'var_prob_s', 'var_sigma_s', 'ks_prob_b', 'ks_prob_h',
       'ks_prob_m', 'ks_prob_s', 'kp_prob_b', 'kp_prob_h', 'kp_prob_m',
       'kp_prob_s']

features = ['hard_hm','hard_hs', 'hard_ms', 'var_prob_b', 'var_sigma_b',
            'var_prob_h', 'var_sigma_h',
        'var_prob_m', 'var_sigma_m', 'var_prob_s', 'var_sigma_s', 'ks_prob_b', 'ks_prob_h',
       'ks_prob_m', 'ks_prob_s', 'kp_prob_b', 'kp_prob_h', 'kp_prob_m',
       'kp_prob_s']

X_df_coord = data[features_coord].dropna()
X_df = X_df_coord[features]
X = X_df.copy().to_numpy()
```

```python
X_df.max()
```

### Aplicar logaritmo y luego normalizaci√≥n.

```python
import pandas as pd
from sklearn import preprocessing

# Log transformation
# But, since we have zeros, we do the transformation log(X + 1)
x = np.log(X_df.var_sigma_b.values + 0.05) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,4] = x_scaled.flatten()

x = np.log(X_df.var_sigma_h.values + 0.05) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,6] = x_scaled.flatten()

x = np.log(X_df.var_sigma_m.values + 0.05) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,8] = x_scaled.flatten()

x = np.log(X_df.var_sigma_s.values + 0.05) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,10] = x_scaled.flatten()



df = pd.DataFrame(X, columns=X_df.columns)
```

```python
df = df.dropna()
X = df.to_numpy()
```

```python
df.max()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,8))

visualizer.fit(df)        # Fit the data to the visualizer
```

```python
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=visualizer.elbow_value_).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
X[:, 4].min()
```

```python
%matplotlib notebook
plt.rcParams['figure.figsize'] = (10, 10)
plt.style.use('ggplot')

# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','blue', 'green', 'black']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], np.log(X[:, 4]+1), c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 4], marker='*', c=colores, s=1000)
```
