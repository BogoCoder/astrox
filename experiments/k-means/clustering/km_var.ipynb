#### K-Means

```python
X_df_out = data[features_out].dropna()
X_df = X_df_out[features]
X = X_df.copy().to_numpy()
```

```python
X_df.max()
```

### Aplicar logaritmo y luego normalizaci√≥n.

```python
# Log transformation

x = np.log(X_df.var_sigma_b.values, where = X_df.var_sigma_b.values >0)  #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,4] = x_scaled.flatten()

x = np.log(X_df.var_sigma_h.values, where = X_df.var_sigma_h.values >0) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,6] = x_scaled.flatten()

x = np.log(X_df.var_sigma_m.values, where = X_df.var_sigma_m.values >0) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,8] = x_scaled.flatten()

x = np.log(X_df.var_sigma_s.values, where = X_df.var_sigma_s.values >0) #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
x_scaled = min_max_scaler.fit_transform(x.reshape(-1,1))
X[:,10] = x_scaled.flatten()

X_df = pd.DataFrame(X, columns=X_df.columns)
X_df = X_df.dropna()
X = X_df.to_numpy()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,8))

visualizer.fit(X_df)        # Fit the data to the visualizer
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3).fit(X_df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
kmeans.n_iter_
```

```python
%matplotlib notebook
plt.rcParams['figure.figsize'] = (10, 10)
plt.style.use('ggplot')

colx = 0
coly = 1
colz = 4
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','blue', 'green']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 4], c=asignar,s=20)
ax.scatter(C[:, 0], C[:, 1], C[:, 4], marker='*', c=colores, s=1000)
ax.set(xlabel = X_df.columns[colx])
ax.set(ylabel = X_df.columns[coly])
ax.set(zlabel = X_df.columns[colz])
```
